```python
%matplotlib inline
import os

import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-muted')
plt.rc('ytick',labelsize=20)

# seaborn-muted

activations = [
    'ELU',
    'ReLU',
    'SELU',
    'Softplus',
]

scores = {
    'CIFAR-10': {
        'Inception Score':{
            'GN-GAN': [6.50, 7.66, 3.43, 1.56],
            'SN-GAN': [2.17, 7.42, 5.93, 1.43],
            'WGAN-GP': [2.09, 7.03, 5.65, 2.38],
        },
        'FID': {
            'GN-GAN': [52.14, 21.69, 122.4, 394.2],
            'SN-GAN': [187.1, 23.71, 50.51, 395.7],
            'WGAN-GP': [229.3, 30.45, 55.76, 273.3],
        },
    },
    'STL-10': {
        'Inception Score':{
            'GN-GAN': [6.92, 8.54, 5.33, 6.83],
            'SN-GAN': [2.11, 8.31, 6.45, 1.36],
            'WGAN-GP': [4.00, 8.56, 6.814, 1.676],
        },
        'FID': {
            'GN-GAN': [98.56, 37.96, 98.67, 73.22],
            'SN-GAN': [206.1, 41.97, 64.45, 290.9],
            'WGAN-GP': [136.5, 44.2, 78.1, 265.9],
        },
    }
}
```

```python
fig, axs = plt.subplots(1, 4, figsize=(40, 10))

fontsize = 40
legend_fontsize = 30
barWidth = 0.2
r1 = np.arange(len(activations))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
tick_center = [x + barWidth for x in r1]

os.makedirs('./figures', exist_ok=True)
for i, score in enumerate(['Inception Score', 'FID']):
    for j, dataset in enumerate(['CIFAR-10', 'STL-10']):
        cnt = i * 2 + j
        ax = axs.flat[cnt]
        ax.bar(r1, scores[dataset][score]['GN-GAN'], width=barWidth, edgecolor='white', label='GN-GAN')
        ax.bar(r2, scores[dataset][score]['SN-GAN'], width=barWidth, edgecolor='white', label='SN-GAN')
        ax.bar(r3, scores[dataset][score]['WGAN-GP'], width=barWidth, edgecolor='white', label='WGAN-GP')
#         ax.set_yticklabels(np.arange(9), fontsize=fontsize)
        ax.set_xticks(tick_center)
        ax.set_xticklabels(activations, fontsize=fontsize, rotation=45)
#         ax.set_yticklabels(ax.get_yticklabels(), fontsize=16)
        ax.set_ylabel(score, fontsize=fontsize)
        ax.set_title(dataset, y=-0.3, fontsize=fontsize)
        if cnt == 3:
            ax.legend(fontsize=legend_fontsize)
fig.savefig('./figures/vis_ablation_activation.png')
```
